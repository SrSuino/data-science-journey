```python
# Bem-vindo
print("Olá, bem vindo!")

# Variáveis simples
nome = input("Digite seu nome: ") #Definindo a variável "nome"
idade = int(input("Digite sua idade: ")) #Definindo a variável "idade"
altura = float(input("Digite sua altura: ")) #Definindo a variável "altura"

print(f"Olá, {nome}! Você tem {idade} anos e mede {altura}m. Daqui a 10 anos, você terá {idade + 10} anos de idade!") #Enviando a mensagem para o terminal (stdout)
```

```python

print(type(nome)) #Testando o tipo da variável "nome"
print(type(idade)) #Testando o tipo da variável "idade"
print(type(altura)) #Testando o tipo da variável "altura"
```
